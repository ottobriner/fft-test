```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# construct time and signal, plot
N = 400
x = np.linspace(0, (2*np.pi) * 4, N)
y = np.sin(x)
plt.scatter(x, y)
```

```python
# compute dft, power, and frequency bins
yfft = np.fft.rfft(y)
power = np.abs(yfft)**2
xfreq = np.fft.rfftfreq(x.shape[-1])
```

```python
# plot power(freq) and show peak at 1/N cycles/step
plt.plot(xfreq, power)
plt.show()
```
